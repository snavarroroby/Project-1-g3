```python
global_pwrplant['primary_fuel'].value_counts()

fuel_list = ['Hydro', 'Wind', 'Gas', 'Coal', 'Oil', 'Biomass',            
'Waste',              
'Nuclear',             
'Geothermal',          
'Storage',             
'Other',               
'Cogeneration',         
'Petcoke',              
'Wave and Tidal']

solar_df = global_pwrplant.loc[global_pwrplant['primary_fuel']=="Solar"]
solar_df

grouped_solar_df = solar_df.groupby('country_long').sum()
grouped_solar_df.sort_values(by=['capacity_mw'], ascending = False).head(5)

# # def group_fuel(fuel_type):
# #     fuel_type_df = global_pwrplant.loc[global_pwrplant['primary_fuel']=={fuel_type}]
# #     grouped_fuel_df = fuel_type_df.groupby('country_long').sum()
# #     print(grouped_fuel_df.sort_values(by=['capacity_mw'], ascending = False).head(5))

global_pwrplant_reduced = global_pwrplant[['country_long', 'capacity_mw', 'primary_fuel']]

global_pwrplant_reduced

def group_fuel(fuel_type):
    fuel_type_df = global_pwrplant_reduced.loc[global_pwrplant_reduced['primary_fuel']==fuel_type]
    grouped_fuel_df = fuel_type_df.groupby('country_long').sum()
    
        
    grouped_fuel_df["primary_fuel"] = fuel_type
    grouped_fuel_df = grouped_fuel_df.sort_values(by=['capacity_mw'], ascending = False)
    
    return grouped_fuel_df.iloc[:5]
    
df_group = group_fuel("Solar")

for fuel_type in fuel_list:
           
    df = group_fuel(fuel_type)
    df_group = pd.concat([df_group,df])
        
print(df_group.to_string())
```

```python
colors = ['red', 'orange', 'yellow', 'lightgreen', 'lightblue']

country_by_fuel = df_group

country_by_fuel_df = country_by_fuel.reset_index()

country_by_fuel_df

solar_fuel = country_by_fuel_df.loc[(country_by_fuel_df['primary_fuel']=='Solar')]
solar_fuel

# Solar Pie Chart
solar_pie = plt.pie(solar_fuel['capacity_mw'], colors=colors, labels=solar_fuel['country_long'],
        autopct='%1.1f%%', pctdistance=0.85)

centre_circle = plt.Circle((0, 0), 0.5, fc='white')
fig = plt.gcf()

fig.gca().add_artist(centre_circle)
plt.title('Top 5 Countries - Solar')

plt.legend(labels=solar_fuel['country_long'], loc="lower left",bbox_to_anchor=(0.85,1.025), title="Countries")
plt.axis('equal')
plt.tight_layout()
plt.show()

#Hydro pie chart

hydro_fuel = country_by_fuel_df.loc[(country_by_fuel_df['primary_fuel']=='Hydro')]
hydro_fuel

# Pie Chart
hydro_pie = plt.pie(hydro_fuel['capacity_mw'], colors=colors, labels=hydro_fuel['country_long'],
         autopct='%1.1f%%', pctdistance=0.85)

centre_circle = plt.Circle((0, 0), 0.5, fc='white')
fig = plt.gcf()

fig.gca().add_artist(centre_circle)
plt.title('Top 5 Countries - Hydro')

plt.legend(labels=hydro_fuel['country_long'], loc="lower left",bbox_to_anchor=(0.85,1.025), title="Countries")
plt.axis('equal')
plt.tight_layout()
plt.show()



#wind chart
wind_fuel = country_by_fuel_df.loc[(country_by_fuel_df['primary_fuel']=='Wind')]
wind_fuel

# Pie Chart
wind_pie = plt.pie(wind_fuel['capacity_mw'], colors=colors, labels=wind_fuel['country_long'],
        autopct='%1.1f%%', pctdistance=0.85)

centre_circle = plt.Circle((0, 0), 0.5, fc='white')
fig = plt.gcf()

fig.gca().add_artist(centre_circle)
plt.title('Top 5 Countries - Wind')

plt.legend(labels=wind_fuel['country_long'], loc="lower left",bbox_to_anchor=(0.85,1.025), title="Countries")
plt.axis('equal')
plt.tight_layout()
plt.show()

# Coal Pie Chart
coal_fuel = country_by_fuel_df.loc[(country_by_fuel_df['primary_fuel']=='Coal')]
coal_fuel


coal_pie = plt.pie(coal_fuel['capacity_mw'], colors=colors, labels=coal_fuel['country_long'],
        autopct='%1.1f%%', pctdistance=0.85)

centre_circle = plt.Circle((0, 0), 0.5, fc='white')
fig = plt.gcf()

fig.gca().add_artist(centre_circle)
plt.title('Top 5 Countries - Coal')

plt.legend(labels=coal_fuel['country_long'], loc="lower left",bbox_to_anchor=(0.85,1.025), title="Countries")
plt.axis('equal')
plt.tight_layout()
plt.show()


# Nuclear Pie Chart
nuclear_fuel = country_by_fuel_df.loc[(country_by_fuel_df['primary_fuel']=='Nuclear')]
nuclear_fuel


nuclear_pie = plt.pie(nuclear_fuel['capacity_mw'], colors=colors, labels=nuclear_fuel['country_long'],
        autopct='%1.1f%%', pctdistance=0.85)

centre_circle = plt.Circle((0, 0), 0.5, fc='white')
fig = plt.gcf()

fig.gca().add_artist(centre_circle)
plt.title('Top 5 Countries - Nuclear')

plt.legend(labels=nuclear_fuel['country_long'], loc="lower left",bbox_to_anchor=(0.85,1.025), title="Countries")
plt.axis('equal')
plt.tight_layout()
plt.show()

#Oil Pie Chart

oil_fuel = country_by_fuel_df.loc[(country_by_fuel_df['primary_fuel']=='Oil')]
oil_fuel


oil_pie = plt.pie(oil_fuel['capacity_mw'], colors=colors, labels=oil_fuel['country_long'],
        autopct='%1.1f%%', pctdistance=0.85)

centre_circle = plt.Circle((0, 0), 0.5, fc='white')
fig = plt.gcf()

fig.gca().add_artist(centre_circle)
plt.title('Top 5 Countries - Oil')

plt.legend(labels=oil_fuel['country_long'], loc="lower left",bbox_to_anchor=(0.85,1.025), title="Countries")
plt.axis('equal')
plt.tight_layout()
plt.show()
```
